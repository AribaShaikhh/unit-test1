```python
#Que-7
plt.figure(figsize=(25,5))
sns.set_style("darkgrid")
sns.countplot(x=df["State"],palette="Set2",edgecolor="blue")
plt.title("State with highest sales",size=35)
plt.xlabel("State",size=15)
plt.ylabel("Amount",size=15)
plt.yticks(size=10)
plt.xticks(size=10)
plt.show()
```

```python
plt.figure(figsize=(25,5))
sns.countplot(x=df["Occupation"],palette="Set3",edgecolor="black")
plt.title("effect of purchase",size=35)
plt.xlabel("Occupation",size=15)
plt.ylabel("Purchase",size=15)
plt.yticks(size=10)
plt.xticks(size=10)
plt.show()
```
